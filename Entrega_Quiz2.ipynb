## Gráficos de líneas

En general, cuando se tienen una o muchas variable numéricas que guardan una relación con una variable independiente y ordenada (como por ejemplo el tiempo), lo ideal es realizar gráficos de líneas. En estos gráficos se pueden representar y comparar fácilmente varias series de datos, así como también mostrar comportamientos asociados a la variable independiente.

Cuando se tienen múltiples valores de la serie por cada valor de la varaiable independiente, se  gráfica de manera automática el intervalo de confianza del 95% calculado usando el método de bootstrapping.

## Gráficos de Distribución

El tipo de gráficas que se van a construir asume que los datos fuente tiene al menos dos variables: la primera es una variable numérica continua o discreta y la segunda es una varaible categórica de interés para la cual se quiere observar la distribución de los valores numéricos comparados por categoría. Cuando se tienen dos variables numéricas continuas es posible generar una de ellas la discretización mediante quantiles y de esta forma analizar su distribuación

La idea es tener una representación de la distribución de los datos por categoría para comparación. Para esto _Seaborn_ presenta dos tipos de gráficos llamados __strip__ y __swarm__.

Distribución de las observaciones
Estos tipos de gráficos permiten analizar la distribución de las observaciones por cada una de las categorías. Seaborn tiene varias opciones para pintar la distribución de los datos como son: diagrama de bigotes, bigotes extendido y violines. Estos diagramas hacen uso del cálculo de los cuartiles de la distribución de los datos para hallar: mediana, quartil 1, quartil 3, extensión de los bigotes inferior y superior a 1.5 IQR. Las muestras que quedan fuera de este rango se marcan como datos atípicos. en los diagramas de caja los resultados corresponden a valores reales.
