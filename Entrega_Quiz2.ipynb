Cuando se tienen dos variables numÃ©ricas continuas es posible generar una de ellas la discretizaciÃ³n mediante quantiles y de esta forma analizar su distribuaciÃ³n

La idea es tener una representaciÃ³n de la distribuciÃ³n de los datos por categorÃ­a para comparaciÃ³n. Para esto _Seaborn_ presenta dos tipos de grÃ¡ficos llamados __strip__ y __swarm__.

Estos tipos de grÃ¡ficos permiten analizar la distribuciÃ³n de las observaciones por cada una de las categorÃ­as. Seaborn tiene varias opciones para pintar la distribuciÃ³n de los datos como son: diagrama de bigotes, bigotes extendido y violines. Estos diagramas hacen uso del cÃ¡lculo de los cuartiles de la distribuciÃ³n de los datos para hallar: mediana, quartil 1, quartil 3, extensiÃ³n de los bigotes inferior y superior a 1.5 IQR. Las muestras que quedan fuera de este rango se marcan como datos atÃ­picos. en los diagramas de caja los resultados corresponden a valores reales.
